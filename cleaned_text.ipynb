```python
import os
import cv2
import pytesseract

# Define the folder containing images
image_folder = r"C:\Users\deekshith\Desktop\Hackathon\images"  # Change this to your folder path
output_folder = r"C:\Users\deekshith\Desktop\Hackathon\output_texts"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get a list of all image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Process each image
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    
    # Read image
    image = cv2.imread(image_path)

    # Convert image to text
    text = pytesseract.image_to_string(image)

    # Save the extracted text
    output_text_path = os.path.join(output_folder, image_file.replace('.png', '.txt').replace('.jpg', '.txt').replace('.jpeg', '.txt'))
    with open(output_text_path, 'w', encoding='utf-8') as f:
        f.write(text)

    print(f"Processed: {image_file}")

print("OCR completed for all images!")
```

```python
sample_text_file = os.path.join(output_folder, os.listdir(output_folder)[0])  # Pick first text file
with open(sample_text_file, 'r', encoding='utf-8') as f:
    sample_text = f.read()

print("📜 Sample Extracted Text:\n")
print(sample_text[:1000])  # Show first 1000 characters
print("SI")
```

```python
import os

# Define path where all text files are stored
output_folder = r"C:\Users\deekshith\Desktop\Hackathon\output_texts"
final_output_file = os.path.join(output_folder, "final_output.txt")

# Merge all text files into one
with open(final_output_file, 'w', encoding='utf-8') as final_file:
    for txt_file in sorted(os.listdir(output_folder)):  # Sort ensures correct order
        if txt_file.endswith(".txt"):  # Only process .txt files
            txt_path = os.path.join(output_folder, txt_file)
            with open(txt_path, 'r', encoding='utf-8') as f:
                final_file.write(f.read() + "\n\n")  # Add text with spacing

print(f"✅ Merged all text files into: {final_output_file}")
```

```python
import re

# Function to clean extracted text
def clean_text(text):
    """Remove extra spaces, special characters, and multiple new lines"""
    text = re.sub(r'\n+', '\n', text)  # Remove multiple new lines
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Remove non-ASCII characters
    text = re.sub(r'[^\w\s.,!?-]', '', text)  # Keep only words, spaces, and punctuation
    return text.strip()

# Read & clean the final output file
with open(final_output_file, 'r', encoding='utf-8') as f:
    cleaned_text = clean_text(f.read())

# Save the cleaned text
cleaned_output_file = os.path.join(output_folder, "cleaned_final_output.txt")
with open(cleaned_output_file, 'w', encoding='utf-8') as f:
    f.write(cleaned_text)

print(f"✅ Cleaned text saved in: {cleaned_output_file}")
```
